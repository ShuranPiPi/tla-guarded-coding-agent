# Provider Fallback Demo

Shows how `AGENT_PROVIDER=auto` chooses OpenAI first and Gemini as fallback based on environment variables.

```python
from pathlib import Path
import os, sys
from dotenv import load_dotenv

ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
sys.path.insert(0, str(ROOT))
load_dotenv(ROOT / '.env')

from agent.llm import available_providers, select_provider, fallback_provider
```

```python
print('Configured providers:', available_providers())
try:
    chosen = select_provider(os.environ.get('AGENT_PROVIDER', 'auto'))
    print('Selected:', chosen)
    print('Fallback:', fallback_provider(chosen))
except Exception as exc:
    print('Provider selection failed:', exc)
```
