# TLC-Spec-Guarded Agent Demo

Runs the package workflow on one sample task. Set `SPEC_MODE` to `example` for the compact finite-example template or `specification` for the structured-specification + low-level TLA template. A real OpenAI or Gemini key is required for live LLM generation.

```python
from pathlib import Path
import json, sys
from dotenv import load_dotenv

ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
sys.path.insert(0, str(ROOT))
load_dotenv(ROOT / '.env')

from agent import run_agent
```

```python
TASK_NAME = 'two_sum'
SPEC_MODE = 'specification'  # 'example' or 'specification'

tasks = json.loads((ROOT / 'tasks' / 'sample_tasks.json').read_text(encoding='utf-8'))
task = next(t for t in tasks if t['name'] == TASK_NAME)
final = run_agent(task, max_spec_retries=2, max_code_retries=1, spec_mode=SPEC_MODE)
final['workflow'], final.get('terminal_reason')
```

```python
print('\n'.join(final['history']))
print('\nSpec mode:', final.get('spec_mode'))
print('\nSpec result:', final.get('spec_result'))
if final.get('structured_spec'):
    print('\nStructured specification:\n', final['structured_spec'][:1200])
print('\nGenerated code:\n', final.get('code', ''))
```
