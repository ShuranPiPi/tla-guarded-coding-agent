# TLC-Spec-Guarded Agent Demo

Runs the current package workflow on one sample task. A real OpenAI or Gemini key is required.

```python
from pathlib import Path
import json, sys
from dotenv import load_dotenv

ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
sys.path.insert(0, str(ROOT))
load_dotenv(ROOT / '.env')

from agent import run_agent
```

```python
tasks = json.loads((ROOT / 'tasks' / 'sample_tasks.json').read_text(encoding='utf-8'))
task = next(t for t in tasks if t['name'] == 'two_sum')
final = run_agent(task, max_spec_retries=2, max_code_retries=1)
final['workflow'], final.get('terminal_reason')
```

```python
print('\n'.join(final['history']))
print('\nSpec result:', final.get('spec_result'))
print('\nGenerated code:\n', final.get('code', ''))
```
